```python
import recipes
recipe = recipes.recipe
```

```python
# food contains names of all 101 categories
i=0
idx_food = {}             # Names of food item
food_idx = {}
for key in recipe:
    idx_food[i] = key
    food_idx[key] = i
    i=i+1
```

```python
import re
import heapq
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
stopwords = set(stopwords.words('english'))
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corpus = []
for key in recipe:
    corpus.append([recipe[key]]) 
```

```python
corpora = []

for item in corpus:
    item[0] = item[0].lower()
    item[0] = re.sub(r"[^a-zA-Z]",' ',item[0])
    item[0] = item[0].split()
    item[0] = [i for i in item[0] if i not in stopwords]
    item[0] = [i for i in item[0] if len(i)>2]
    corpora.append(item[0])
```

```python
from gensim.models import KeyedVectors
w2v = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
recipe_text= {}
for key in recipe:
    recipe_text[key] = corpora[food_idx[key]]
```

```python
def get_vectors(first_map,second_map):
    first_vec = dict()
    for uid,content in first_map.items():
        temp=[]
        for element in content:
            try:
                temp.append(second_map[element])
            except KeyError:
                pass
        first_vec[uid] = np.mean(temp,axis=0)
    return first_vec
```

```python
document_vec = get_vectors(recipe_text,w2v)
```

```python
document_vec['samosa'].shape
```

```python
recommended = []
for i in range(0,101):
    x={}
    for j in range(0,101):
        q = cosine_similarity(document_vec[idx_food[j]].reshape(1,-1),document_vec[idx_food[i]].reshape(1,-1))
        x[j] = q[0][0]
    recommended.append(x)
    
recommended
```

```python
import operator

recommended_all_num = []
for i in range(0,101):
    x =recommended[i]
    sorted_x = sorted(x.items(), key=operator.itemgetter(1))
    sorted_x.reverse()
    recommended_all_num.append(sorted_x[1:5])
recommended_all_num
```

```python
recommended_name = []
for item in recommended_all_num:
    temp = []
    for i in item:
        temp.append(idx_food[i[0]])
    recommended_name.append(temp)
```
